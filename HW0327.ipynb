## 04-2. matplotlib 라이브러리 자유자재로 사용하기

### 기초 그래프 그리기

```python
#1. tips 데이터 집합을 불러와 tips에 저장하기
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
#2. 히스토그램 : 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도로 사용
# 데이터프레임의 열을 변수라고 부름
# 일변량 그래프 : 변수를 하나만 사용해서 그린 그래프
```

```python
#3. 기본 틀(fig)를 마련하고 그래프 격자(axes)를 구성
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```python
#4. hist를 이용해 히스토그램 만들기. bins를 이용해 x축 간격 조정
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig
```

```python
#5. 산점도 : 변수 2개를 사용해서 만드는 그래프
# 이변량 그래프 : 변수 2개를 사용해서 그린 그래프프
```

```python
#6. 산점도 그래프 그리기
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
#7. 박스 그래프 : 이산형 변수와 연속형 변수를 함께 사용하는 그래프
#이산형 변수 : 성별과 같이 명확하게 구분되는 값
#연속형 변수 : tip과 같이 명확하게 셀 수 없는 범위의 값값
```

```python
#8. boxplot을 이용해 박스 그래프 그리기
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
               labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

### 다변량 그래프 그리기

```python
#1.여성인 경우 0을 반환하고 남성인 경우 1을 반환하는 함수
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

```python
#2. recode_sex 함수가 반환한 값을 데이터 프레임에 열을 새로 생성하여 추가한 것
#apply를 이용하여 sex 열에 recode_sex 함수를 브로드캐스팅
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
#3. scatter 메서드에 s,c 인잣값으로 테이블당 인원 수와 성별의 치환값을 전달한 것
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10,
    c = tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

##04-3. seaborn 라이브러리 자유자재로 사용하기

### 단변량 그래프 그리기 - 히스토그램

```python
#1. 데이터 불러오기 및 패키지 import
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
#2. subplots, distplot을 이용하여 히스토그램 그리기
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
#3. distpot을 이용하여 히스토그램과 밀집도 그래프 그리기
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
#4. 밀집도 그래프만 나타내려면 hist 인자를 False로 지정
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
#5. 여러 그래프를 한 번에 출력하기
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
#6. count 그래프 그리기
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

### 다변량 종류의 이변량 그래프 그리기

```python
#1. seaborn 라이브러리로 산점도 그래프 그리기

ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data = tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#2.jointplot을 이용해 산점도 그래프와 히스토그램 그리기
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
#3. 육각 그래프 : 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
#4. kdeplot을 이용해 이차원 밀집도 그리기
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                 data2=tips['tip'],
                 shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

##04-5. seaborn 라이브러리 그래프 스타일 설정하기

### 그래프에 스타일 적용하기

```python
fig, ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
#2. whitegrid로 그래프 배경 설정하기
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
#3. for문을 이용해 모든 스타일을 하나씩 적용한 그래프 그리기
fig = plt.figure( )

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
